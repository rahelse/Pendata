###  4. Hitung Entropy Awal (sebelum split)

$$
Entropy(D) = -p_1 \log_2(p_1) - p_2 \log_2(p_2)
$$

$$
p_1 = \frac{8}{15},\quad p_2 = \frac{7}{15}
$$

$$
Entropy(D) = -\frac{8}{15} \log_2 \frac{8}{15} - \frac{7}{15} \log_2 \frac{7}{15} \approx 0.996
$$

###  5. Entropy setelah split

Group Left (≤ 152.5):

* Data: No 3, 11, 4, 14, 1, 12, 5, 2 → 8 Data
* Kelas: 8 Sus verrucosus, 0 Sus scrofa

$$
Entropy(left) node
* P1 =   -\frac{8}{8} = 1         
* P2 =   -\frac{0}{8} = 0
$$

$$
Entropy(left) = -\ 1.\log_2 \ (1)- 0.log_2 (0) = 0
$$

Group Right (> 152.5):

* Data: No 7, 9, 15, 6, 8, 13. 10 → 7 data
* Kelas: 7 Sus scrofa, 0 Sus verrucosus

$$
Entropy(right) node
* P1 =   -\frac{0}{7} = 0         
* P2 =   -\frac{7}{7} = 1
$$

$$
Entropy(right) = -\ 0.\log_2\ (0)- 1.log_2 (1) = 0
$$

Information Gain:

$$
IG = Entropy(D) - \left( \frac{8}{15} \ .0  + \frac{7}{15} \ .0 \right) \approx 0.996 - 0 \approx \boxed{0.996}
$$

###  6. Gini Index

$$
Gini(D)  = 1 - \ (p_1^2 + p_2^2)
$$
$$
P1 = 1-\left((\frac{8}{15})^2(\frac{7}{15})^2\right)
$$
$$
= 1- \left ( \frac{64}{225}\  + \frac{49}{225} \ \right)
$$
$$
= 1- \left ( \frac{113}{225}\right)
$$
$$
= 1- \left ( \frac{112}{225}\right) \approx {0.498}
$$

Group Left (≤ 155):

* 8 Sus verrucosus, 1 Sus scrofa

$$
Gini(left) = 1 - \left(\frac{8}{9}\right)^2 - \left(\frac{1}{9}\right)^2 = 1 - 0.7901 - 0.0123 = 0.1976
$$

Group Right (> 155):

* 6 Sus Scrofa, 0 Sus verrucosus

$$
Gini(right) = 1-(0)^2-(1)^2 = 0
$$

Gini Split:

$$
Gini_{split} = \frac{9}{15} \cdot 0.1976 + \frac{6}{15} \cdot 0 = \boxed{0.1186}
$$

## Visualisasi Decision Tree

Ini adalah code program untuk mencari decision tree sekaligus rules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Babi ===
data = {
    "Panjang Tubuh (cm)": [138, 140, 142, 144, 145, 147, 148, 150, 155, 158, 159, 160, 162, 163, 165],
    "Berat Tubuh (kg)":   [37, 39, 42, 44, 45, 47, 48, 50, 55, 58, 59, 60, 62, 63, 65],
    "Kelas": [
        "Sus verrucosus", "Sus verrucosus", "Sus verrucosus", "Sus verrucosus", "Sus verrucosus",
        "Sus verrucosus", "Sus verrucosus", "Sus verrucosus",
        "Sus scrofa", "Sus scrofa", "Sus scrofa", "Sus scrofa", "Sus scrofa", "Sus scrofa", "Sus scrofa"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # scrofa=1, verrucosus=0

# === Fitur dan Target ===
X = df[["Panjang Tubuh (cm)", "Berat Tubuh (kg)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)
```

## Visualisasi Splitting

```python
uji_baru = pd.DataFrame({
    'Panjang Tubuh (cm)': [135, 155],   # contoh: satu di bawah rentang, satu di tengah cluster scrofa
    'Berat Tubuh (kg)': [36, 60]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (p, b) in enumerate(zip(uji_baru['Panjang Tubuh (cm)'], uji_baru['Berat Tubuh (kg)'])):
    print(f"Data {i+1}: Panjang={p}, Berat={b} → Prediksi: {hasil[i]}")

# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "rb"
plot_step = 0.5

x_min, x_max = X['Panjang Tubuh (cm)'].min() - 5, X['Panjang Tubuh (cm)'].max() + 5
y_min, y_max = X['Berat Tubuh (kg)'].min() - 5, X['Berat Tubuh (kg)'].max() + 5

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Panjang Tubuh (cm)', 'Berat Tubuh (kg)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Panjang Tubuh (cm)'], X.iloc[idx]['Berat Tubuh (kg)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

# Plot titik uji baru
plt.scatter(uji_baru['Panjang Tubuh (cm)'], uji_baru['Berat Tubuh (kg)'],
            c='yellow', edgecolor='k', s=100, marker="*", label='Uji Baru')

plt.xlabel("Panjang Tubuh (cm)")
plt.ylabel("Berat Tubuh (kg)")
plt.title("Decision Boundary (Panjang vs Berat) - Data Babi")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
